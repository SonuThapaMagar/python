```python
""" 
Create a class Book with attributes title, author, and pages. Add a method display() that prints 
the book details.
"""
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def display(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Pages: {self.pages}")


book1 = Book("Atomic Habits", "James Clear", 246)
book1.display()
```

```python
"""
Create a class Rectangle with attributes length and width. Include methods to calculate area 
and perimeter. 
"""
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


rect = Rectangle(10, 5)

print("Area:", rect.area())
print("Perimeter:", rect.perimeter())
```

```python
"""
Write a class Car with attributes make, model, year, and speed. Add methods accelerate() 
(increase speed by 10) and brake() (decrease speed by 10). 
"""
class Car:
    def __init__(self, make, model, year, speed):
        self.make = make
        self.model = model
        self.year = year
        self.speed = speed

    def accelerate(self):
        self.speed += 10

    def brake(self):
        self.speed -= 10


car = Car("Toyota", "Corolla", 2024, 0)

car.accelerate()
print(car.speed)

car.accelerate()
print(car.speed)

car.brake()
print(car.speed)
```

```python
""" 
Create a class BankAccount with attributes account_number and balance. Include methods 
deposit(amount) and withdraw(amount). 
"""
class BankAccount:

    def __init__(self,acc_no,balance=0):
       self.acc_no=acc_no
       self.balance=balance

    def deposit(self,amount):
        print(f"Your current balance is:{self.balance}")
        self.balance+=amount
        print(f"After deposit, your current balance is:{self.balance}")

    def withdraw(self,withdraw_amount):
        print(f"Your current balance is:{self.balance}")
        if self.balance<withdraw_amount:
            print("Insufficient balance")
        else:
            self.balance-=withdraw_amount
        print(f"After withdraw, your current balance is:{self.balance}")

account = BankAccount("123456", 5000)

account.deposit(2000)
print(account.balance)

account.withdraw(3000)
print(account.balance)
```

```python
"""
Define a class Student with attributes name, age, and grades (list). Add a method to calculate
average grade.
"""
class Student:
    def __init__(self, name, age, grades):
        self.name = name
        self.age = age
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)


student = Student("Sonu", 22, [80, 90, 85, 95])

print("Average:", student.average_grade())
```

```python
"""
Create a class Employee with attributes name, salary, and department. Add a method
give_raise(percentage).
"""
class Employee:
    def __init__(self, name, salary, department):
        self.name = name
        self.salary = salary
        self.department = department

    def give_raise(self, percentage):
        self.salary += self.salary * (percentage / 100)


employee = Employee("Ram", 50000, "IT")

employee.give_raise(10)

print(employee.salary)
```

```python
"""
Write a class Circle with attribute radius. Include methods to calculate area and
circumference. (Use 3.14 for π)
"""
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def circumference(self):
        return 2 * 3.14 * self.radius


circle = Circle(5)

print(f"Area: {circle.area():.2f}")
print(f"Circumference: {circle.circumference():.2f}")
```

```python
"""
Define a class Laptop with attributes brand, ram, and price. Add a method
discount(percentage) that returns the discounted price.
"""
class Laptop:
    def __init__(self, brand, ram, price):
        self.brand = brand
        self.ram = ram
        self.price = price

    def discount(self, percentage):
        return self.price - (self.price * percentage / 100)


laptop = Laptop("Dell", "16GB", 100000)

print("Discounted Price:", laptop.discount(15))
```

```python
"""
Create a class Counter with an attribute count. Include methods increment(), decrement(),
and reset().
"""
class Counter:
    def __init__(self):
        self.count = 0

    def increment(self):
        self.count += 1

    def decrement(self):
        self.count -= 1

    def reset(self):
        self.count = 0


counter = Counter()

counter.increment()
counter.increment()
print(counter.count)

counter.decrement()
print(counter.count)

counter.reset()
print(counter.count)
```
