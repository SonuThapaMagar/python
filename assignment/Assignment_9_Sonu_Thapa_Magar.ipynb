```python
# Use map() and lambda to double all numbers in this list: [1, 2, 3, 4, 5].
my_list=[1,2,3,4,5]
result=list(map(lambda number:number **2,my_list))
print(f"List after doubling all the numbers in the list : {result}")
```

```python
#Use map() and lambda to convert this list of strings to uppercase: ["apple", "banana", "cherry"]
fruits=["apple", "banana", "cherry"]
result=list(map(lambda fruit: fruit.upper(),fruits))
print(f"List after converting to uppercase using lambda:{result} ")
```

```python
#Write a recursive function to find the factorial of a number and trace the output in a copy as well.
def factorial(number):
    if number==1:
      return 1
    return number*factorial(number-1)

number=int(input("Enter a number to find the factorial: "))
result=factorial(number)
print(f"Factorial of {number} is :{result}")
```

```python
#Write a recursive function to find the length of a string and trace the output as well.
def find_length(user_str):
    if user_str == "":
        return 0
    return 1 + find_length(user_str[1:])

user_input=input("Enter a string to find its length: ")
result=find_length(user_input)
print(f"The length of the string is: {result}")
```
