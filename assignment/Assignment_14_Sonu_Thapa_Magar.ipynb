```python
"""Q1. Create a class Engine with horsepower and a start() method.
Create a Car class that uses Composition (contains an Engine object).
Add a start_car() method
"""
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

    def start(self):
        print(f"Engine with {self.horsepower} HP started.")

class Car:
    def __init__(self, brand,engine):
        self.brand = brand
        self.engine=engine

    def start_car(self):
        print(f"Starting the {self.brand}...")
        self.engine.start()
```

```python
engine=Engine(150)
car=Car("Toyota",engine)
```

```python
car.start_car()
```

```python
"""
Q2. Abstract class Notifier(ABC) with send(message, recipient)
• Implement EmailNotifier, SMSNotifier, PushNotifier
• User class that contains multiple Notifiers (Composition)
• notify_user(message) method that uses all attached notifiers
"""
from abc import ABC,abstractmethod

class Notifier(ABC):
    @abstractmethod
    def send(self, message, recipient):
        pass

class EmailNotifier(Notifier):
    def send(self, message, recipient):
        print(f"Sending Email to {recipient}: {message}")

class SMSNotifier(Notifier):
    def send(self, message, recipient):
        print(f"Sending SMS to {recipient}: {message}")

class PushNotifier(Notifier):
    def send(self, message, recipient):
        print(f"Sending Push Notification to {recipient}: {message}")

class User:
    def __init__(self,recipient, name):
        self.name = name
        self.recipient = recipient
        self.notifiers = []

    def add_notifier(self, notifier):
        self.notifiers.append(notifier)

    def notify_user(self, message):
        print(f"\nSending notifications to {self.name}")
        for notifier in self.notifiers:
            notifier.send(message, self.recipient)
```

```python
user1 = User("Sonu", "sonu@gmail.com")

user1.add_notifier(EmailNotifier())
user1.add_notifier(SMSNotifier())
user1.add_notifier(PushNotifier())

user1.notify_user("Your order has been delivered.")
```

```python
"""
Q3. Create an abstract class PaymentProcessor(ABC) with an
abstract method process_payment(amount).
Implement CreditCardProcessor and PayPalProcessor.
Create a ShoppingCart class that contains a PaymentProcessor
(Composition).
"""
from abc import ABC, abstractmethod

class PaymentProcessor(ABC):
    @abstractmethod
    def process_payment(self,amount):
        pass

class CreditCardProcessor(PaymentProcessor):
    def process_payment(self, amount):
        print(f"Processing credit card payment of ${amount}.")

class PayPalProcessor(PaymentProcessor):
    def process_payment(self, amount):
        print(f"Processing PayPal payment of ${amount}.")

class ShoppingCart:
    def __init__(self, payment_processor):
        self.payment_processor = payment_processor

    def checkout(self, amount):
        print("Processing Payment...")
        self.payment_processor.process_payment(amount)
```

```python
credit=CreditCardProcessor()
```

```python
cart=ShoppingCart(credit)
cart.checkout(100)
```

```python
paypal=PayPalProcessor()
cart2=ShoppingCart(paypal)
cart2.checkout(200)
```

```python
"""
Q4. Implement the concept of access modifiers (public, protected, and
private) in the class Bank: Assume attributes by yourself and
implement the concept of name mangling to view private attributes
"""
class Bank:

    def __init__(self, bank_name, manager_name, balance):
        self.bank_name = bank_name          # Public
        self._manager_name = manager_name   # Protected
        self.__balance = balance            # Private

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited Rs.{amount}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn Rs.{amount}")
        else:
            print("Insufficient Balance")

    def display_balance(self):
        print(f"Current Balance: Rs.{self.__balance}")
```

```python
bank1 = Bank("Nepal Bank", "Ram", 50000)

# Public
print("Public Attribute:")
print(bank1.bank_name)
```

```python

# Protected
print("Protected Attribute:")
print(bank1._manager_name)
```

```python
# Private through method
bank1.display_balance()

bank1.deposit(5000)
bank1.display_balance()

bank1.withdraw(10000)
bank1.display_balance()
```
