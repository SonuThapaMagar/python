```python
#Lists
#Addition
fruits =["Apple","Banana","Mango","Orange","Papaya"]
print("Elements in a list: ",fruits)
```

```python
#append()
fruits.append("Litchi")
print("Elements after appending: ",fruits)
```

```python
#insert(index,value)
fruits.insert(5,"Watermelon")
print("Elements after inserting: ",fruits)
```

```python
#extend
vegetables=["Potato","Tomato"]
fruits.extend(vegetables)
print(fruits)
```

```python
#Remove in Lists
#remove()
fruits.remove('Apple')
print("Elements after removing: ",fruits)
```

```python
#del[index]
del fruits[0]
print("Elements after deleting from index 0: ",fruits)
```

```python
#pop()
fruits.pop()
print("Elements after popping out without giving index: ",fruits)
fruits.pop(1)
print("Elements after popping out with index: ",fruits)
```

```python
#clear()
fruits.clear()
print("Elements after using clear(): ", fruits)
```

```python
#Dictionary
students={"name":"Sonu","age":23,"university":"aeu","email":"sonumagar787@gmail.com"}
print(students)
print(students.items())
print(students.keys())
print(students.values())
print(students.get("name"))
```

```python
#Changing 
students["name"]="Incessant Rain Studio"
print("After changing value: ",students)
```

```python
#Adding 
students["departments"]="3D-Frontend"
print("Adding in dictionary: ",students)
```

```python
#Updating in dictionary
services={"subject":"Python","faculty":"IT"}
students.update(services)
print("Updating dictionary in students: ",students)
```

```python
#Removing using pop(),popitem()
students.pop("age")
print("After popping age: ",students)
```

```python
#Removing using popitem()
students.popitem()
print(students)
```

```python
# Removing using del
del students["email"]
print(students)
```

```python
#clear()
students.clear()
print(students)
```

```python
#Nested Lists
students=[["Ram",20,"A"],["Shyam",23,"B"],["Sita", 21, "B"],
    ["Hari", 22, "A"]]
print(students)
print(students[0])
```

```python
#Accessing Ram's age
print("Accessing Ram's age: ",students[0][1])
```

```python
#Access Hari's grade
print("Accessing Hari's grade: ",students[3][2])
```

```python
#Changing Sita's age
students[2][1]=19
print("After changing Sita's age: ",students[2][1])
print(students)
```

```python
#Adding new student
students.append(["Gita",21,"A+"])
print(students)
```

```python
#inserting a student
students.insert(1,["Krishna",20,"B"])
print("After inserting student: ",students)
```

```python
#Adding a subject to Shyam
students[2].append("Python")
print("After adding subject to Shyam: ",students)
```

```python
#Popping the student
students.pop()
print("Popping the last student: ",students)
```

```python
#Removing the second student
students.pop(1)
print("Popping the second student: ",students)
```

```python
#Remove using remove()
students.remove(["Hari",22,'A'])
print(students)
```

```python
#Delete using del
del students[0]
print(students)
```

```python
#Nested Dictionaries
students = {
    "101": {
        "name": "Ram",
        "age": 20,
        "grade": "A"
    },
    "102": {
        "name": "Sita",
        "age": 21,
        "grade": "B"
    }
}
print(students["101"]["name"])
```

```python
#Accessing Sita's age
print(students["102"]["age"])
```

```python
#Modifying values of Ram
students["101"]["grade"]="A+"
students
```

```python
#Adding new student
students["103"]={
    "name":"Sonu",
    "age":23,
    "grade":"A+"
}
students
```

```python
#Adding new key
students["103"]["address"]="Bhaisepati"
students
```

```python
#Removing a key
students["101"].pop("grade")
print("After popping grade key: ",students)
```

```python
#Removing student
students.pop("101")
students
```

```python
#Deleting using del
del students["102"]["grade"]
students
```

```python
#Deleting entire dictionary
del students["102"]
students
```
