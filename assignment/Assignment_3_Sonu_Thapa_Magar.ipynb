```python
#Reversing a list using slicing and swapping first and last element
numbers=[10,20,45,56,67,78,89,90]

print(numbers)
```

```python
#Reversing list using slicing
reversed=numbers[::-1]
print("List after reversing: ",reversed)
```

```python
#Swapping first and last elements
reversed[0],reversed[-1]=reversed[-1],reversed[0],
print("After swapping first and last elements in a list: ",reversed)
```

```python
#2nd largest element in a list
a = [10, 90, 70, 50, 60, 20]
a.sort(reverse=True)
print(a)
print("Second largest element: ",a[1])
```

```python
#Unique elements in a list
a = [1, 2, 3, 5, 6, 7, 8, 9]
b = [1, 3, 5, 7, 9]

unique=list(set(a) ^ set(b))
print("Unique elemets in a set: ",unique)
```

```python
#Remove an Item and Append It to the End
numbers= [5, 10, 15, 20]

numbers.append(numbers.pop(2))
print("List after removing item at index 2 and appending at the end: ",numbers)
```

```python
#Appending number at the end
store1 = {'apple': 5, 'banana': 3}
store2 = {'orange': 4, 'banana': 10}

merged_store=store1 | store2
print("Merged dictionary is: ",merged_store)
```

```python
# count
nums = (1, 5, 3, 5, 2, 5, 4)

count=nums.count(5)
first_index=nums.index(5)
print(f"5 appears {count} time and first index is {first_index}")
```

```python
#unpacking
fruits = ("apple", "banana", "orange", "berries")
start,*rest,end=fruits

print(f"First unpacked variable value is {start}, last is {end} and rest of them are {rest}")
```
