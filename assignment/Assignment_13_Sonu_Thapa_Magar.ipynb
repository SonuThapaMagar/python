```python
#Hierarchical Inheritance
class A: #Parent class
    def __init__(self,pname):
        self.pname=pname

    def display(self):
        return f"Hi, I am parent {self.pname}"
    
class B(A): #Child1 Class
    def __init__(self, pname,c1name):
        super().__init__(pname)
        self.c1name=c1name

    def display(self):
        p_display=super().display() #A.display(self)
        print(p_display)
        return f"Hi, I am first child {self.c1name}"

class C(A): #Child2 Class
    def __init__(self, pname,c2name):
        super().__init__(pname)
        self.c2name=c2name

    def display(self):
        p_display=super().display() #A.display(self)
        print(p_display)
        return f"Hi, I am second child {self.c2name}"
```

```python
obj1 = B("Mom", "Son")
print(obj1.display())
```

```python
obj2 = C("Mom", "Daughter")
print(obj2.display())
```

```python
class A:
    def __init__(self, pname):
        self.pname = pname

    def info(self):
        return f"Parent name is {self.pname}"


class B(A):
    def __init__(self, pname, bname):
        A.__init__(self,pname)
        self.bname = bname

    def display_b(self):
        return f"First child name is {self.bname}"


class C(A):
    def __init__(self, pname, cname):
        A.__init__(self,pname)
        self.cname = cname

    def display_c(self):
        return f"Second child name is {self.cname}"


class D(B, C):
    def __init__(self, pname, bname, cname, dname):
        B.__init__(self, pname, bname)  
        C.__init__(self, pname, cname)   
        self.dname = dname

    def display_d(self):
        return f"Grandchild name is {self.dname}"
```

```python
obj = D("Father", "Son1", "Son2", "Grandson")
print(obj.info())
print(obj.display_b())
print(obj.display_c())
print(obj.display_d())
```

```python
class A:
    def __init__(self, a):
        self.a = a

    def show_a(self):
        return f"Value from A: {self.a}"


class B:
    def __init__(self, b):
        self.b = b

    def show_b(self):
        return f"Value from B: {self.b}"


class C(A):
    def __init__(self, a, c):
        A.__init__(self, a)
        self.c = c

    def show_c(self):
        return f"Value from C: {self.c}"


class D(B):
    def __init__(self, b, d):
        B.__init__(self, b)
        self.d = d

    def show_d(self):
        return f"Value from D: {self.d}"


class E(C, D):
    def __init__(self, a, b, c, d, e):
        C.__init__(self, a, c)
        D.__init__(self, b, d)
        self.e = e

    def show_e(self):
        return f"Value from E: {self.e}"
```

```python
obj = E("A Value", "B Value", "C Value", "D Value", "E Value")
print(obj.show_a())
print(obj.show_b())
print(obj.show_c())
print(obj.show_d())
print(obj.show_e())
```

```python
"""
Create a base class Shape :
__init__(self, name)
area(self) method (return 0 by default)
info(self) method that returns name only
Create a subclass Rectangle that inherits from Shape:
Add attributes: length and width
Override area() to calculate and return the rectangle area (length ×
width)
Override info() to display name, dimensions, and area
"""
class Shape:
    def __init__(self,name):
        self.name=name

    def area(self):
        return 0

    def info(self):
        return f"Shape Name: {self.name}"

class Rectangle(Shape):
    def __init__(self, name,length,width):
        super().__init__(name)
        self.length=length
        self.width=width

    def area(self):
        return self.length *self.width

    def info(self):
        return (
            f"Shape Name: {self.name}\n"
            f"Length: {self.length}\n"
            f"Width: {self.width}\n"
            f"Area: {self.area()}"
        )
```

```python
rect=Rectangle("Rectangle",5,3)
print(rect.info())
```

```python
"""
Create two base classes:
Flyable with fly() method
Swimmable with swim() method
Create class AmphibiousVehicle that inherits from both Flyable and
Swimmable
Add name and speed attributes
Override both fly() and swim() methods with specific behavior
Add perform_actions() method that calls both overridden methods
"""
class Flyable:
    def fly(self):
        return "This object can fly."

class Swimmable:
    def swim(self):
        return "This object can swim."

class AmphibiousVehicle(Flyable, Swimmable):
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed

    def fly(self):
        return f"{self.name} is flying at {self.speed} km/h."

    def swim(self):
        return f"{self.name} is swimming at {self.speed} km/h."

    def perform_actions(self):
        print(self.fly())
        print(self.swim())
```

```python

vehicle = AmphibiousVehicle("SeaPlane", 300)
vehicle.perform_actions()
```
