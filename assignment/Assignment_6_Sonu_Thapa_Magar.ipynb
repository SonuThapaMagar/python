```python
# Write a Python program that calculates a student's total marks, percentage, and final
# letter grade based on their scores in Math, Science, English, History, and Art.
subjects=[ "Math", "Science", "English", "History",  "Art"]
total_marks=0

for subject in subjects:
    marks = int(input(f"Enter the marks for {subject}: "))
    total_marks += marks

# Calculate percentage
percentage=(total_marks /500)*100

if percentage >= 90:
    grade = "A"
elif percentage >= 80:
    grade = "B"
elif percentage >= 70:
    grade = "C"
elif percentage >= 60:
    grade = "D"
else:
    grade = "F"

#Results
print("---Result---")
print(f"Total marks: {total_marks}")
print(f"Total percentage (%): {percentage}")
print(f"Grade: {grade}")
```

```python
# Write a Python program that prints numbers from 1 to 100.
for i in range(1,101):
    if i%3==0 and i%5==0:
        print("FizzBuzz")
    elif i%3==0:
        print("Fizz")
    elif i%5==0:
        print("Buzz")
    print(i)
```

```python
#Write a Python program that asks the user to enter a number and prints its multiplication
#table from 1 to 10, except for even numbers, using a loop.

num=int(input("Enter the number to generate the multiplication table"))

j=1
print(f"Multiplication table of {num}")
while j<=10:
    print(f"{num}x{j}={num *j}")
    j+=1
```

```python
#Write a Python program that stores the secret number as 25.
secret_number = 25

guess = int(input("Guess the number: "))

while guess != secret_number:
    if guess < secret_number:
        print("Too Low!")
    else:
        print("Too High!")

    guess = int(input("Guess again: "))

print("Congratulations! You guessed the correct number.")
```
