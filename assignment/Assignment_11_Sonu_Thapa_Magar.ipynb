```python
"""1. Create a Rectangle class with:
● Attributes: length, width
● Methods:
○ area()
○ perimeter()
- Create 3 Rectangle objects. One object displays area() and the other
object displays perimeter()
"""
class Rectangle:
   
    def __init__(self,length,width):
        self.length=length
        self.width=width

    def area(self):
        result=self.length*self.width
        return result
    
    def perimeter(self):
        result=2*(self.length+self.width)
        return result
    
length = int(input("Enter the length of Rectangle 1: "))
width = int(input("Enter the width of Rectangle 1: "))

length2 = int(input("Enter the length of Rectangle 2: "))
width2 = int(input("Enter the width of Rectangle 2: "))

obj_rectangle_1=Rectangle(length,width)
obj_rectangle_2=Rectangle(length2,width2)

#Result
print("=====Rectangle 1=====")
print("Length:", obj_rectangle_1.length)
print("Width:", obj_rectangle_1.width)
print("Area:", obj_rectangle_1.area())
print("Perimeter:", obj_rectangle_1.perimeter())

print()

print("=====Rectangle 2=====")
print("Length:", obj_rectangle_2.length)
print("Width:", obj_rectangle_2.width)
print("Area:", obj_rectangle_2.area())
print("Perimeter:", obj_rectangle_2.perimeter())
```

```python
"""
Create a Person class with:
● Attributes: name, age
● Methods:
○ introduce() → prints "Hi, I am [name], and I am [age] years old."
Create 2 Person objects and call introduce() on both.
"""
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def introduce(self):
         return f"Hi, I am {self.name}, and I'm {self.age} years old."
    
name1=input("Enter the name: ")
age1=input("Enter the age: ")

name2=input("Enter the name: ")
age2=input("Enter the age: ")

obj1_person=Person(name1,age1)
obj2_person=Person(name2,age2)

#Result
print("====Result===")
print(obj1_person.introduce())
print(obj2_person.introduce())
```

```python
"""
Create a Book class with:
● Attributes: title, author
● Methods:
○ details() → prints "Book: [title] by [author]"
Create 2 Book objects and show their details
"""
class Book:
    def __init__(self,title,author):
        self.title=title
        self.author=author
    
    def details(self):
        return f"Book: {self.title} by {self.author}"
    
book_title_1=input("Enter the book title: ")
book_author_1=input("Enter the book author: ")

book_title_2=input("Enter the book title: ")
book_author_2=input("Enter the book author: ")

obj1_book=Book(book_title_1,book_author_1)
obj2_book=Book(book_title_2,book_author_2)

print("======Result=====")
print(obj1_book.details())
print(obj2_book.details())
```

```python
"""Create a Circle class with:
● Attribute: radius
● Methods:
○ area() → returns area (use 3.14 for π)
○ circumference() → returns circumference
Create one circle with radius 5 and print area and circumference.
"""
class Circle:

    def __init__(self,radius):
        self.radius=radius
    
    def area(self):
        return 3.14 * self.radius * self.radius
    
    def circumference(self):
        return 2 * 3.14 * self.radius
    
circle=Circle(5)

print("=====Result=====")
print("Radius:", circle.radius)
print(f"Area: {circle.area():.2f}")
print(f"Circumference: {circle.circumference():.2f}")
```
