```python
# capitalize() : Makes the first character uppercase
str="learning python in incessant rain studio"
print(str.capitalize())
```

```python
#casefold() :Converts to lowercase for case-insensitive comparisons.
str="Converts to lowercase for case-insensitive comparisons."
print(str.casefold())
```

```python
#center(): Centers the string within a given width.
str="AI/ML Training"
print(str.center(20,'*'))
```

```python
#count():Counts occurrences of a substring.
str="Artificial Intelligence"
print(str.count('i'))
```

```python
#encode():Encodes the string into bytes.
str="power"
print(str.encode())
```

```python
#endswith(): Checks whether the string ends with a suffix.
str="Day1.ipynb"
print(str.endswith('.ipynb'))
```

```python
#expandtabs(): Replaces tab characters with spaces.
str="Name\tAge\t"
print(str.expandtabs(4))
```

```python
#find(): Returns first index of a substring or -1.
str="The power of positive thinking"
print(str.find('v'))
```

```python
#format() : formats value into a string
name="Sonu"
course="Python"
message="Hello {}, Welcome to {} programming".format(name,course)
print(message)
```

```python
#format_map(): Formats using a mapping/dictionary.
student={
    "name":"Sonu",
    "course":"Python"
}
print("Name: {name}, Course: {course}".format_map(student))
```

```python
#index(): Find the first position of a substring.
email="sonumagar787@gmail.com"
print(email.find("@"))
```

```python
#isalnum(): Check if a string contains only letters and numbers.
username="Sonu23"
print(username.isalnum())
```

```python
#isalpha():  True if all characters are letters.
name="Sonu"
print(name.isalpha())
```

```python
#isascii(): True if all characters are ASCII.
str="Sonu"
print(str.isascii())
```

```python
#isdecimal(): True if all are decimal digits.
year = "2083"

print(year.isdecimal())
```

```python
#isdigit(): True if all are digits
year="2083"
print(year.isdigit())
```

```python
#isidentifier(): Checks valid Python identifier
str="student$"
print(str.isidentifier())
```

```python
#islower(): Checks all letters are lowercase. lower(): converts to lowercase
str="Sonu Thapa Magar"
print(str.islower())
print(str.lower())
```

```python
#isnumeric(): Checks numeric characters.
str1="Sonu Thapa Magar"
print(str1.isnumeric())

str="Sonu Thapa Magar"
print(str.isnumeric())
```

```python
#isprintable(): Checks printable characters.
text = "Hello World"
print(text.isprintable())
```

```python
#isspace(): Checks whitespace only.
text = "     "

print(text.isspace())
```

```python
#istitle(): Checks title case.
book = "Rich Dad Poor Dad"

print(book.istitle())
```

```python
#isupper(): Checks uppercase. upper():converts to uppercase
str="python"
print(str.isupper())
print(str.upper())
```

```python
#join(): Joins iterable using separator.
skills = ["HTML", "CSS", "JavaScript", "Python"]

print(", ".join(skills))
```

```python
#ljust()
str="Loading"
print(str.ljust(15,"."))
```

```python
#lstrip(): Removes leading spaces.
str="   Python"
print(str.lstrip())
```

```python
#maketrans(): Creates translation table.
table=str.maketrans("sonu","1234")
print(table)
```

```python
#partition(): Splits at first separator.
email="sonumagar787@gmail.com"
print(email.partition("@"))
```

```python
#removeprefix():Removes matching prefix
url="https://www.python.org/"
print(url.removeprefix("https://"))
```

```python
#removesuffix(): Removes matching suffix.
url="google.com"
print(url.removesuffix(".com"))
```

```python
#replace(): Replaces substring
str="I like java"
print(str.replace("java","python"))
```

```python
#rfind(): Finds last occurrence.
path = "python/Day1.ipynb"
print(path.rfind("/"))
```

```python
#rindex(): Last index or error.
path = "python/Day1.ipynb"
print(path.rindex("/"))
```

```python
#rjust(): Right-aligns string.
price = "250"
print(price.rjust(8, "0"))
```

```python
#rpartition(): Splits at last separator.
str="python/Day1.pynb"
print(str.rpartition("/"))
```

```python
#rsplit(): Splits from right.
path="Users/USER/Desktop/python/Day.ipynb"
print(path.rsplit("/",1))
```

```python
#rstrip(): Removes trailing spaces.
text = "Python      "
print(text.rstrip())
```

```python
#split(): Splits into list.
str = "Python Java JavaScript React"
print(str.split())
```

```python
#splitlines(): Splits at line breaks
text = "Apple\nBanana\nOrange"
print(text.splitlines())
```

```python
#startswith(): Checks starting prefix.
url="https.google.com"
print(url.startswith("https"))
print(url.startswith("www"))
```

```python
#strip(): Removes leading/trailing spaces.
text = "    Python Programming    "
print(text.strip())
```

```python
#swapcase(): Swaps upper/lower case.
str="AI/ML is awesome"
print(str.swapcase())
```

```python
#title(): Capitalizes each word.
str="Python Basics"
print(str.title())
```

```python
#translate(): Translates characters.
table=str.maketrans("aeiou","56789")
print(table)
```

```python
#zfill(): Pads with leading zeros.
id="23"
print(id.zfill(6))
```
