```python
'''
Write a Python program using asyncio similar to the factorial example.
Requirements:
● Create an async function called sum_of_squares(name, number)
● It should calculate the sum of squares from 1² + 2² + ... + number²
● Print a message in each iteration like: "Task {name}: Computing square of {i}..."
● Use await asyncio.sleep(1) in each loop iteration to simulate work
● After finishing, print the final result: "Task {name}: sum of squares({number}) = {result}"
● In main(), run three tasks concurrently for numbers 3, 5, and 6 using asyncio.gather()
● Print the final list of results
'''
```

```python
import asyncio  # python async library
```

```python
#async function
async def sum_of_squares(name,number):
    result=0
    for i in range(1,number+1):
        print(f"Task {name}: Computing square of {i}...")
        await asyncio.sleep(1) #pauses 1 sec and run another async func
        result += i**2
    print(f"Task {name}: sum of squares({number}) = {result}")
    return result
```

```python
#main function
async def main():
    results=await asyncio.gather(sum_of_squares("A",3),sum_of_squares("B",5),sum_of_squares("C",6)) # runs all task together
    print(f"Final results: {results}")
```

```python
await main() #async function call
```

```python
"""
Write a Python program using asyncio:
Requirements:
● Create an async function called multiplication_table(name, number)
● It should print the multiplication table for numbers from 1 to 5 (e.g., 5 x 1 = 5, 5x 2 = 10,
etc.).
● Print a message in each step like: "Task {name}: Calculating {number} x {i}..."
● Use await asyncio.sleep(0.8) in each iteration
● After finishing, print: "Task {name}: Completed multiplication table for {number}"
● In main(), run three tasks concurrently for numbers 4, 6, and 8 using asyncio.gather()
● At the end, print "All tables completed!"
"""
```

```python
import asyncio
```

```python
async def multiplication_table(name, number): #async function
  for i in range(1,6):
    print(f"Task {name}: Calculating {number} X {i}...")
    await asyncio.sleep(0.8)
    print(f"{number} X {i}={number * i}")

  print(f"Task {name}: Completed multiplication table for {number}")
```

```python
async def main(): #async main function
  await asyncio.gather(multiplication_table("A",4),multiplication_table("B",6),multiplication_table("C",8))
  print("All tables completed!")
```

```python
await main() # async function call
```

```python
"""
Write a decorator named execution_time that prints 'Function Started' before execution and
'Function Completed' after execution. Apply it to calculate_square(number), call the function,
and display the returned result.
"""
```

```python
def execution_time(func): #decorator function
  def inner(number): #inner function as decorator returns function

    print('Function Started')
    result=func(number) # calls original function
    print("Function Completed")
    return result
  return inner # returns function
```

```python
@execution_time
def calculate_square(number): # original function
  return number **2
```

```python
result=calculate_square(3) #function call
print("Returned Result:", result)
```
