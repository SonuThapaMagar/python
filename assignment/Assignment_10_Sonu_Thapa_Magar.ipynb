```python
"""
1. Write a function named calculate_area(length, width)
that calculates and returns the area of a rectangle.
Requirements:
1. Accept length and width as parameters;
2. Return the area;
3. Call the function and display the returned value.
2. Write a function display_employee(**kwargs) that
accepts employee information using keyword arguments
and displays each key-value pair.
# Take at least 5 employee information by yourself.
"""

def calculate_area(length,width):
   try:
        if length<=0 or width<=0:
            raise ValueError("Length and width must be greater than 0")
        
        area=length * width
        return area
   except Exception as e:
      print(type(e).__name__)

try:
   l=float(input("Enter the length: "))
   w=float(input("Enter the width: "))

   result=calculate_area(l,w)
   print("Area of rectangle is: ",result)
except Exception as e:
    print("Enter numberic value only")
```

```python
"""
Write a function display_employee(**kwargs) that
accepts employee information using keyword arguments
and displays each key-value pair.
# Take at least 5 employee information by yourself.
"""
def display_employee(**details):
    try:
        for key,value in details.items():
            print(f"{key}:{value}")
            print("-"*30)
    except Exception as e:
        print(type(e).__name__)  
          
print("======Employee Details====== ")

display_employee(name="Sonu",id=101,department="Frontend");
display_employee(name="Alex",id=123,department="Backend");
display_employee(name="Ram",id=234,department="QA");
display_employee(name="Shyam",id=456,department="Devops");
display_employee(name="Hari",id=678,department="UI/UX");
```

```python
"""
Implement the concept of a function in Assignment 6.

Question 1: Student Grade Calculator
Write a Python program that calculates a student's total marks, percentage, and final
letter grade
based on their scores in Math, Science, English, History, and Art.
Requirements:
1. Prompt the user to enter the marks for all five subjects individually.
2. Convert each input into an integer.
3. Calculate the total marks obtained (out of 500).
4. Calculate the overall percentage.
5. Assign a grade: A (90-100), B (80-89.99), C (70-79.99), D (60-69.99), F (below 60).
6. Display Total Marks, Percentage (2 decimal places), and Final Grade.
"""

def get_marks():
    subjects=[ "Math", "Science", "English", "History",  "Art"]
    total_marks=0

    try:
        for subject in subjects:
            marks = int(input(f"Enter the marks for {subject}: "))
            if marks < 0 or marks > 100:
                raise ValueError("Marks must be between 0 and 100.")

            total_marks += marks
        return total_marks
    except ValueError:
        print("Error: Enter valid marks in numeric value")
        return None

def calculate_percentage(total):
    return (total / 500) * 100

def calculate_grade(percentage):
    if percentage >= 90:
        return "A"

    elif percentage >= 80:
        return "B"

    elif percentage >= 70:
        return "C"

    elif percentage >= 60:
        return "D"

    else:
        return "F"

def display_result(total, percentage, grade):
    print("\n------ Result ------")
    print("Total Marks:", total)
    print(f"Percentage: {percentage:.2f}%")
    print("Grade:", grade)


total = get_marks()

if total is not None:
    percentage = calculate_percentage(total)
    grade = calculate_grade(percentage)
    display_result(total, percentage, grade)
```

```python
"""
#Write a Python program to print all even numbers between 1 and 100.
"""
def check_even():
    try:
        for num in range(1,101):
            if num%2==0:
                print(num)
    except Exception as e:
        print(type(e).__name__)


check_even()
```

```python
'''
Write a Python program that prints numbers from 1 to 100.
- Print "Fizz" for multiples of 3.
- Print "Buzz" for multiples of 5.
- Print "FizzBuzz" for multiples of both.
- Otherwise, print the number.
'''

def check_division():
    try:
        for num in range(1,101):
            if num%3==0 and num%5==0:
                print("FizzBuzz")
            elif num%3==0:
                print("Fizz")
            elif num%5==0:
                print("Buzz")
            else:
                print(num)
    except Exception as e:
        print(type(e).__name__)
check_division()
```

```python
'''
Write a Python program that asks the user to enter a number and prints its multiplication
table from 1 to 10, except for even numbers, using a loop.
'''
def mul_table(num):
    try:
       for i in range(1,11):
            if i%2!=0:
                print(f"{num}x{i}={num*i}")
    except Exception as e:
        print(type(e).__name__)        
try:
    num=int(input("Enter a number :"))
    mul_table(num)
except ValueError:
    print("Enter numeric value only")
```

```python
'''
Write a Python program that stores the secret number as 25.
- Ask the user to guess the number.
- Keep asking until the correct number is entered.
- If the guess is too low, print "Too Low!"
- If the guess is too high, print "Too High!"
- When the user guesses correctly, print "Congratulations! You guessed the correct
number."
'''
secret_number = 25

def check_guess(num):
    if num < secret_number:
        print("Too Low!")
        return False
    elif num > secret_number:
        print("Too High!")
        return False
    else:
        print("Congratulations! You guessed the correct number.")
        return True

while True:
    try:
        guess = int(input("Guess the secret number: "))
        if check_guess(guess):
            break
    except ValueError:
        print("Invalid input! Please enter a valid integer.")
```
