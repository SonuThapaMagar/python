```python
class MarkerIndustry:
    name="Camlin"
    height=12
    radius=4

    def __init__(self,color):
        self.color=color # color is obj attr
    
    def display(self):
        print(f"Name:{self.name},height={self.height}")
```

```python
m_blue=MarkerIndustry("blue")
print(m_blue)
```
