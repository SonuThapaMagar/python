```python
class MarkerIndustry:
    name="Camlin"
    height=12
    radius=4

    def __init__(self,color):
        self.color=color # color is obj attr

    @classmethod
    def change_height(cls,new_height):
        cls.height=new_height

    @classmethod
    def change_name(cls,new_name):
        cls.name=new_name

    @classmethod
    def change_radius(cls,new_radius):
        cls.radius=new_radius
        
    @staticmethod
    def info():
        print("This is Marker Industry")

    @staticmethod
    def calc_sum(x,y):
        return x+y

    def display(self):
        print(f"Name:{self.name},height={self.height}, radius={self.radius},color={self.color}")
```

```python
obj1=MarkerIndustry("blue")
```

```python
obj1.change_height(10)
obj1.display()
```

```python
obj1.change_name("Kotlin")
obj1.display()
```

```python
obj1.change_radius(6)
obj1.display()
```

```python
obj1.info()
```

```python
obj1.calc_sum(1,2)
```

```python
class Bank:
    bank_name="Nabil"
    location="Bhaisepati"

    def __init__(self,acc_name,acc_no,balance=0):
        self.acc_name=acc_name
        self.acc_no=acc_no
        self.balance=balance

    @property
    def email(self):
        return f'{self.acc_name}@gmail.com"'

    def deposit(self,amount):
        print(f"Your current balance is:{self.balance}")
        self.balance+=amount
        print(f"After deposit, your current balance is:{self.balance}")

    def withdraw(self,withdraw_amount):
        print(f"Your current balance is:{self.balance}")
        if self.balance<withdraw_amount:
            print("Insufficient balance")
        else:
            self.balance-=withdraw_amount
        print(f"After withdraw, your current balance is:{self.balance}")

    @classmethod
    def change_location(cls,new_location):
        cls.location=new_location

    def display(self):
        print(f"Bank Name: {self.bank_name}, Location: {self.location}, Account Name: {self.acc_name}, Account Number: {self.acc_no} Email:{self.email}")
```
