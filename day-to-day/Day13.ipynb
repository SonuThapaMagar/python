```python
#OOP 4 main concepts
# Abstraction
# Encapsulation
# Inheritance
# Polymorphism

#Abstraction
class Chatbot:
    def __init__(self,name): #constructor
        self.name=name
        self.login=False
    
    def start_chat(self):
        self.login=True
        return f"Welcome ! {self.name}"
    
    def end_chat(self):
        self.login=False
        return f"Thank you ! {self.name}"
```

```python
obj=Chatbot("Sonu")
obj.name
```

```python
obj.start_chat()
```

```python
obj.end_chat()
```

```python
#Inheritance

#Single Inheritance
class A: #Parent class
    def __init__(self,pname):
        self.pname=pname

    def display(self):
        return f"Hi, I am parent {self.pname}"
    
class B(A): #Child Class
    def __init__(self, pname,cname):
        super().__init__(pname)
        self.cname=cname

    def display(self):
        p_display=super().display() #A.display(self)
        print(p_display)
        return f"Hi, I am child {self.cname}"
```

```python
p1=B("Dog","Puppy")
p1.cname #child name
```

```python
p1.pname #parent name
```

```python
p1.display()
```

```python
#MultiLevel Inheritance
class A: #Parent class
    def __init__(self,pname):
        self.pname=pname

    def display(self):
        return f"Hi, I am parent {self.pname}"
    
class B(A): #Child Class
    def __init__(self, pname,cname):
        super().__init__(pname)
        self.cname=cname

    def display(self):
        p_display=super().display() #A.display(self)
        print(p_display)
        return f"Hi, I am child {self.cname}"
    
class C(B): #Grand child class
    def __init__(self, pname, cname,gcname):
        super().__init__(pname, cname)
        super().__init__(pname, cname)
        self.gcname=gcname

    def display(self):
        p_display=super().display()
        print(p_display)
        return f"Hi, I am grand child {self.gcname}"
```

```python
p1=C("Dog","Puppy","Grandpuppy")
p1.gcname #grandchild name
```

```python
p1.cname # child name
```

```python
p1.pname #parent name
```

```python
p1.display()
```

```python
#Multiple Inheritance
class A: #First Parent class
    def __init__(self,p1name):
        self.p1name=p1name

    def display(self):
        return f"Hi, I am first parent {self.p1name}"

class B: #Second Parent Class
    def __init__(self, p2name):
        self.p2name=p2name

    def display(self):
        return f"Hi, I am second parent {self.p2name}"
    
class C(A,B): #Child class
    def __init__(self, p1name, p2name,cname):
        A.__init__(self,p1name)
        B.__init__(self,p2name)
        self.cname=cname

    def display(self):
        p1_display=A.display(self)
        p2_display=B.display(self)

        print(p1_display)
        print(p2_display)

        return f"Hi, I am child {self.cname}"
```

```python
c_obj=C("Mom","Dad","Daughter")
c_obj.cname
```

```python
print(C.mro())
```

```python
c_obj.p1name
```

```python
c_obj.p2name
```

```python
c_obj.display()
```

```python
#Hierarchical Inheritance
class A: #Parent class
    def __init__(self,pname):
        self.pname=pname

    def display(self):
        return f"Hi, I am parent {self.pname}"
    
class B(A): #Child1 Class
    def __init__(self, pname,c1name):
        super().__init__(pname)
        self.c1name=c1name

    def display(self):
        p_display=super().display() #A.display(self)
        print(p_display)
        return f"Hi, I am first child {self.c1name}"

class C(A): #Child2 Class
    def __init__(self, pname,c2name):
        super().__init__(pname)
        self.cname=c2name

    def display(self):
        p_display=super().display() #A.display(self)
        print(p_display)
        return f"Hi, I am second child {self.c2name}"
```

```python
c_obj=B("Mom","Son")
c_obj.pname
```

```python
c_obj.c1name
```

```python
print(C.mro())
```

```python
c_obj=C("Daughter")
c_obj.c2name
```

```python
c_obj.display()
```
