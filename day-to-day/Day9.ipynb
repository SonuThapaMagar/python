```python
#Recursion
def text_rec(text):
    if len(text)==0:
        print("Done")
    else:
        print(text)
        return text_rec(text[:-1])

text_rec("Hello")
```

```python
def num_rec(a,b):
    if b==1:
        return a 
    else:
        return a+num_rec(a,b-1)
    
num_rec(3,4)
```

```python
#Factorial
def factorial(num):
    fact=1
    for i in range(1,num+1):
        fact=fact*i
    return fact
factorial(5)
```
