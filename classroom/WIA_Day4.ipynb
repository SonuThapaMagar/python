
| Precedence | Operator(s) | Description | Associativity |
| :--- | :--- | :--- | :--- |
| **1 (Highest)** | `()` | Parentheses | **Left-to-Right** |
| **2** | `**` | Exponentiation | **Right-to-Left** |
| **3** | `+x`, `-x`, `~x` | Unary plus, unary minus, and bitwise NOT | **Right-to-Left** |
| **4** | `*`, `/`, `//`, `%` | Multiplication, division, floor division, and modulus | **Left-to-Right** |
| **5** | `+`, `-` | Addition and subtraction | **Left-to-Right** |
| **6** | `<<`, `>>` | Bitwise left and right shifts | **Left-to-Right** |
| **7** | `&` | Bitwise AND | **Left-to-Right** |
| **8** | `^` | Bitwise XOR | **Left-to-Right** |
| **9** | `\|` | Bitwise OR | **Left-to-Right** |
| **10** | `==`, `!=`, `>`, `>=`, `<`, `<=`, `is`, `is not`, `in`, `not in` | Comparisons, identity, and membership operators | **Chained / Non-associative** |
| **11** | `not` | Logical NOT | **Right-to-Left** |
| **12** | `and` | Logical AND | **Left-to-Right** |
| **13 (Lowest)** | `or` | Logical OR | **Left-to-Right** |


```python
x = 3 ** 2 * 2 + 10 // 3 - 4
```

```python
x = 101 % 10 + 2 * 6 // 2 - 2
```

```python
x = (14 + 6) % 7 * 3 // 2 + 1
```
