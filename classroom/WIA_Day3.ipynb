```python
# Set Joints
s1 = {0,1,2,3,4}
s2 = {0,4,5,6,7,8}
print(s1.union(s2))
print(s1)
print(s2)
```

```python
# Set Joints
s1 = {0,1,2,3,4}
s2 = {0,4,5,6,7,8}
print(s1.update(s2))
print(s1)
print(s2)
```

```python
s1 = {1,2,3,4}
s2=  {1,5,4,3}
print(s1 | s2 )
print(s1)
print(s2)
```

```python
## Intersection
s1 = {1,2,3,4,5,6}
s2 = {1,3,5,7,9,0}
print(s1.intersection(s2))
```

```python
print(s1.intersection_update(s2))
s1
```

```python
print(s1 & s2)
```

```python
s1 = {1,2,3,4,5,6}
s2 ={3,4,6,7,8}
print(s1.difference(s2))
s1
```

```python
print(s1.difference_update(s2))
```

```python
print(s1-s2)
```

```python
# symmetric difference
s1 = {1,2,3,4,5,6}
s2 = {0,1,3,5,7,9,11}
print(s1.symmetric_difference(s2))
```

```python
s1.symmetric_difference_update(s2)
print(s1)
```

```python
print(s1 ^ s2)
```
