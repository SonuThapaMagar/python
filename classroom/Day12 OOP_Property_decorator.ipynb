```python
class MarkerIndustry:
  name= "Camlin"
  height = 12
  radius = 1
  def __init__(self,color):
    self.color = color

  @classmethod
  def change_height(cls, new_height):
    cls.height = new_height

  @staticmethod
  def info():
    print("This is Camlin Industry")

  @staticmethod
  def calc_sum(a,b):
    return a+b

  def display(self):
    print (f"Name : {self.name}, Height :{self.height}, Radius:{self.radius},Color:{self.color}")

```

```python
m11 = MarkerIndustry("Pink")
m11.display()
```

```python
m11.calc_sum(2,3)
```

```python
m11.change_height(10)
m11.display()
```

```python
m12 = MarkerIndustry("Green")
m12.display()
```

```python
class Bank:
  bank_name = "Laxmi"
  bank_location = "Chabahil"

  def __init__(self, name,acc,bal=0):
    self.name =  name
    self.acc = acc
    self.bal  = bal

  @property
  def email(self):
    return f"{self.name}@gmail.com" # eq = self.email = Ram@gmail.com

  @email.setter
  def email(self,new_name):
    self.name = new_name
    return f"{self.name}@gmail.com"

  def deposit(self, amt):
    print(f"Your current balance is {self.bal}")
    self.bal+=amt
    print(f"After deposit your balance is {self.bal}")

  def withdraw(self,w_amt):
    if self.bal < w_amt:
      print(f"Insuff Balance")
      print(f"Available balance:{self.bal}")
    else:
      self.bal-=w_amt
      print(f"After withdraw your ava amt is: {self.bal}")

  @classmethod
  def change_loc(cls,n_loc):
    cls.bank_location = n_loc
```
