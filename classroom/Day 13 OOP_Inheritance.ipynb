```python
class MarkerIndustry:
  name= "Camlin"
  height = 12
  radius = 1
  def __init__(self,color):
    self.color = color

  @classmethod
  def change_height(cls, new_height):
    cls.height = new_height

  @staticmethod
  def info():
    print("This is Camlin Industry")

  @staticmethod
  def calc_sum(a,b):
    return a+b

  def display(self):
    print (f"Name : {self.name}, Height :{self.height}, Radius:{self.radius},Color:{self.color}")

```

```python
m11 = MarkerIndustry("Pink")
m11.display()
```

```python
m11.calc_sum(2,3)
```

```python
m11.change_height(10)
m11.display()
```

```python
m12 = MarkerIndustry("Green")
m12.display()
```

```python
class Bank:
  bank_name = "Laxmi"
  bank_location = "Chabahil"

  def __init__(self, name,acc,bal=0):
    self.name =  name
    self.acc = acc
    self.bal  = bal

  @property
  def email(self):
    return f"{self.name}@gmail.com" # eq = self.email = Ram@gmail.com

  @email.setter
  def email(self,new_name):
    self.name = new_name
    return f"{self.name}@gmail.com"

  def deposit(self, amt):
    print(f"Your current balance is {self.bal}")
    self.bal+=amt
    print(f"After deposit your balance is {self.bal}")

  def withdraw(self,w_amt):
    if self.bal < w_amt:
      print(f"Insuff Balance")
      print(f"Available balance:{self.bal}")
    else:
      self.bal-=w_amt
      print(f"After withdraw your ava amt is: {self.bal}")

  @classmethod
  def change_loc(cls,n_loc):
    cls.bank_location = n_loc
```

```python
### Abstraction
class Chatbot:
  def __init__(self,name):
    self.name = name
    self.login = False

  def start_chat(self):
    self.login = True
    return f"Welcome ! {self.name}"

  def end_chat(self):
    self.login = False
    return f"Thank you ! {self.name}"
```

```python
# inheritance 1. Single Inheritance
class A: # parent class
  def __init__(self,pname):
    self.pname = pname

  def display(self):
    return f"Hi I am parent {self.pname}"

class B(A): #child class
  def __init__(self, pname, cname):
    super().__init__(pname) # A.__init__(self,pname)
    self.cname = cname

  def display(self):
    p_display = super().display() # A.display(self)
    print(p_display)
    return f"Hi I am child {self.cname}"



```

```python
b = B("Doggy","puppy")
b.cname
```

```python
b.display()
```

```python
b.display()
```

```python
b.pname
```

```python
# inheritance 2. Multilevel Inheritance
class A: # parent class
  def __init__(self,pname):
    self.pname = pname

  def display(self):
    return f"Hi I am parent {self.pname}"

class B(A): #child class
  def __init__(self, pname, cname):
    super().__init__(pname) # A.__init__(self,pname)
    self.cname = cname

  def display(self):
    p_display = super().display() # A.display(self)
    print(p_display)
    return f"Hi I am child {self.cname}"

class C(B):
  def __init__(self,pname,cname,gcname):
    super().__init__(pname,cname)
    self.gcname = gcname

  def display(self):
    info = super().display()
    print(info)
    return f"I am GrandChild {self.gcname}"


```

```python
c = C("Doggy","Puppy","Tommy")
c.pname
```

```python
c.cname
```

```python
c.gcname
```

```python
c.display()
```

```python
class A: # parent A # multiple
  def __init__(self,aname):
    self.aname=  aname
  def display(self):
    return f"Hi I am {self.aname}"

class B: # parent B
  def __init__(self,bname):
    self.bname=  bname
  def display(self):
    return f"Hi I am {self.bname}"

class C(A,B):
  def __init__(self, aname, bname,cname):
    A.__init__(self,aname)
    B.__init__(self,bname)
    self.cname = cname

  def display(self):
    a_display = A.display(self)
    b_display = B.display(self)
    print(a_display)
    print(b_display)
    return f"Hi I am {self.cname}"

```
