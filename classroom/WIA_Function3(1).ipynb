```python
def calc_factorial(n):
  result = 1
  for i in range(1,n+1):
    result*=i
  return result
```

```python
calc_factorial(5)
```

```python
# recursion
def text_rec(text):
  if len(text)==0:
    return "Done"
  else:
    print(text)
    return text_rec(text[:-1])
```

```python
a=text_rec('Hel')
a
```

```python
def rec_mul(a,b):
  if b ==1:
    return a
  else:
    return a+rec_mul(a,b-1)
```

```python
rec_mul(2,3)
```

```python
a = lambda x: x**2
a(2)
```

```python
def sqrt(n):
  return n**2
```

```python
l= [1,2,3,4]
a = list(map(sqrt,l))
a
```
