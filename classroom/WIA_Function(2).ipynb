```python
import time
```

```python
def petrol_pump(v_id):
  print(f"Turn of {v_id}")
  time.sleep(2)
  print(f"Filling Petrol:{v_id}")
  time.sleep(5)
  print(f"Paid cash:{v_id}")
```

```python
def main():
  start_time = time.time()
  petrol_pump(1023)
  petrol_pump(1480)
  end_time = time.time()
  print(f"Total execution:{end_time-start_time:.2f}sec")
```

```python
main()
```

```python
import asyncio
```

```python
async def petrol_pump(v_id):
  print(f"Turn of {v_id}")
  await asyncio.sleep(2)
  print(f"Filling Petrol:{v_id}")
  await asyncio.sleep(5)
  print(f"Paid cash:{v_id}")
```

```python
async def main():
  start_time = time.time()
  await asyncio.gather(petrol_pump(1023),petrol_pump(1480))
  end_time = time.time()
  print(f"Total execution:{end_time-start_time:.2f}sec")
```

```python
await main()
```

```python
def decorator_func(main):
  def inner():
    main()
    print("World")
  return inner
```

```python
@decorator_func
def main():
  print('Hello')
```

```python
main()
```

```python
def decorator_sum(func):
  def inner(a,b):
    a+=5
    sum = func(a,b)
    return sum
  return inner
```

```python
@decorator_sum
def calc_sum(a,b):
  return a+b
```

```python
calc_sum(2,3)
```
