```python
def calc_sum(a,b,c=6):
  sum = a+b+c
  return sum
```

```python
def user_info(*names):
  print(type(names))
  print(names)
  for name,id in names:
    print(f"User name:{name},User_id:{id}")
```

```python
user = user_info(("Sonu",10),("Binda",9),("Bandana",8))
user
```

```python
def user_info(**info):
  print(info)
  print(type(info))
  print(f"User name: {info["name"]}, Age:{info["age"]}, college:{info["college"]}")
  for k,v in info.items():
    print(f"{k}:{v}")
```

```python
a = user_info(name = "Deepa",age = 23,college= "PK")
a
```

```python
x = 5
def scope():
  ab = 4
  print(f"Inside function x ",x)
  print(f"Inside function ab ",ab)
print(f"Outside ab",ab)
print(f"Outside x", x)
```

```python
scope()
```

```python
x = 5
def glo_scope():
  global ab
  ab = 4
  print(f"Inside function x ",x)
  print(f"Inside function ab ",ab)
print(f"Outside ab",ab)
print(f"Outside x", x)
```

```python
glo_scope()
```
