```python
class MarkerIndustry:
  name= "Camlin"
  height = 12
  radius = 1
  def __init__(self,color):
    self.color = color

  @classmethod
  def change_height(cls, new_height):
    cls.height = new_height

  @staticmethod
  def info():
    print("This is Camlin Industry")

  @staticmethod
  def calc_sum(a,b):
    return a+b

  def display(self):
    print (f"Name : {self.name}, Height :{self.height}, Radius:{self.radius},Color:{self.color}")

```

```python
# inheritance 1. Single Inheritance
class A: # parent class
  def __init__(self,pname):
    self.pname = pname

  def display(self):
    return f"Hi I am parent {self.pname}"

class B(A): #child class
  def __init__(self, pname, cname):
    super().__init__(pname) # A.__init__(self,pname)
    self.cname = cname

  def display(self):
    p_display = super().display() # A.display(self)
    print(p_display)
    return f"Hi I am child {self.cname}"



```

```python
b = B("Doggy","puppy")
b.cname
```

```python
b.display()
```

```python
b.pname
```

```python
# inheritance 2. Multilevel Inheritance
class A: # parent class
  def __init__(self,pname):
    self.pname = pname

  def display(self):
    return f"Hi I am parent {self.pname}"

class B(A): #child class
  def __init__(self, pname, cname):
    super().__init__(pname) # A.__init__(self,pname)
    self.cname = cname

  def display(self):
    p_display = super().display() # A.display(self)
    print(p_display)
    return f"Hi I am child {self.cname}"

class C(B):
  def __init__(self,pname,cname,gcname):
    super().__init__(pname,cname)
    self.gcname = gcname

  def display(self):
    info = super().display()
    print(info)
    return f"I am GrandChild {self.gcname}"


```

```python
c = C("Doggy","Puppy","Tommy")
c.pname
```

```python
c.cname
```

```python
c.gcname
```

```python
c.display()
```

```python
class A: # parent A # multiple
  def __init__(self,aname):
    self.aname=  aname
  def display(self):
    return f"Hi I am {self.aname}"

class B: # parent B
  def __init__(self,bname):
    self.bname=  bname
  def display(self):
    return f"Hi I am {self.bname}"

class C(A,B):
  def __init__(self, aname, bname,cname):
    A.__init__(self,aname)
    B.__init__(self,bname)
    self.cname = cname

  def display(self):
    a_display = A.display(self)
    b_display = B.display(self)
    print(a_display)
    print(b_display)
    return f"Hi I am {self.cname}"

```

```python
# hybrid Inheritance
class A:
  def __init__(self,aname):
    self.aname = aname
  def display(self):
    return f"I am {self.aname}"

class B(A):
  def __init__(self,aname,bname):
    A.__init__(self,aname)
    self.bname = bname

  def display(self):
    a_display = A.display(self)
    print(a_display)
    return f"I am {self.bname}"

class C(A):
  def __init__(self,aname,cname):
    A.__init__(self,aname)
    self.cname = cname

  def display(self):
    a_display = A.display(self)
    print(a_display)
    return f"I am {self.cname}"

class D(B,C):
  def __init__(self,aname,bname,cname,dname):
    B.__init__(self,aname,bname)
    C.__init__(self,aname,cname)
    self.dname = dname

  def display(self):
    b_display = B.display(self)
    c_display = C.display(self)
    return f"I am {self.dname}"

```

```python
d = D("Anna","Bell","Chris","Dina")
print(D.mro())
```

```python
d.bname
```

```python
# Composition

class Customer:
  def __init__(self, name, location):
    self.name = name
    self.location = location

  def c_info(self):
    return f"{self.name} from {self.location}"

class Product:
  def __init__(self,item,price):
    self.item = item
    self.price = price

  def p_info(self):
    return f"has ordered {self.item}, Costing {self.price}"

class Order:
  def __init__(self,name,location,item,price,order_id):
    self.c1 = Customer(name,location)
    self.p1 = Product(item,price)
    self.order = order_id

  def display_info(self):
    print(self.c1.c_info())
    print(self.p1.p_info())
    return f"Order id is {self.order}"
```

```python
o2 = Order("Alish","Bhaktapur","Tripod",2500,234)
o2.display_info()
```

```python
c1=  Customer("Alish","Bhaktapur")
c1.c_info()
```

```python
# aggregation
class Employee:
  def __init__(self, name,emp_id):
    self.name=  name
    self.id = emp_id

  def e_info(self):
    return f"Employee name {self.name} with id {self.id}"

class Department:
  def __init__(self,dname,obj_e):
    self.dname=  dname
    self.obj_e = obj_e
  def d_info(self):
    print(self.obj_e.e_info())
    return f"Department name is {self.dname}"
```

```python
e1 = Employee("Nisha",1232)
d1 = Department("IT",e1)
```

```python
d1.d_info()
```

```python
# Abstract class Method
from abc import ABC,abstractmethod

class Payment(ABC):
  @abstractmethod
  def payment(self):
    pass

class Esewa(Payment):
  def pay(self):
    return "Payment made through Esewa"
  def payment(self):
    return "Payment made through Esewa"

class Khalti(Payment):
  def payment(self):
    return "Payment made through Khalti"
```
